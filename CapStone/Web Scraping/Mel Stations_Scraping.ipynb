```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import csv
import json

print('Libraries imported.')
source = requests.get('https://en.wikipedia.org/wiki/List_of_railway_stations_in_Melbourne').text
soup = BeautifulSoup(source, 'lxml')
```

```python
gdp_table = soup.find("table", attrs={"class": "wikitable"})
gdp_table_data = gdp_table.tbody.find_all("a")
```

```python
stations_list=[]
lat=[]
lon=[]
gdp_table = soup.find("table", attrs={"class": "wikitable"})
gdp_table_data = gdp_table.tbody.find_all("a")
for list in gdp_table_data:
    station=(list.contents)
    stations_list.append(station[0])
    source_station = requests.get("https://en.m.wikipedia.org/"+list['href']).text
    try:
        lat.append(BeautifulSoup(source_station, 'lxml').find("span", attrs={"class": "latitude"}).contents[0])
    except:
        lat.append(np.nan)
    try:
        lon.append(BeautifulSoup(source_station, 'lxml').find("span", attrs={"class": "longitude"}).contents[0])
    except:
        lon.append(np.nan)
```

```python
print(len(lat))
print(len(stations_list))
```

```python
latitude=[];
longitude=[];
for i in range(0,len(lat)):
    try:
        latitude.append(-(float(lat[i].split("′")[0].split("°")[0])+float(lat[i].split("′")[0].split("°")[1])/60+
              (float(lat[i].split("″")[0].split("′")[1]))/3600))
    except Exception:
        try:
            latitude.append(-(float(lat[i].split("′")[0].split("°")[0])+
                  (float(lat[i].split("″")[0].split("′")[1]))/3600)) 
        except Exception:
            try:
                latitude.append(-(float(lat[i].split("′")[0].split("°")[0])+float(lat[i].split("′")[0].split("°")[1])/60)) 
            except:
                latitude.append(np.nan)
    try:
        longitude.append((float(lon[i].split("′")[0].split("°")[0])+float(lon[i].split("′")[0].split("°")[1])/60+
              (float(lon[i].split("″")[0].split("′")[1]))/3600))
    except Exception:
        try:
            longitude.append((float(lon[i].split("′")[0].split("°")[0])+
                  (float(lon[i].split("″")[0].split("′")[1]))/3600)) 
        except Exception:
            try:
                longitude.append((float(lon[i].split("′")[0].split("°")[0])+float(lon[i].split("′")[0].split("°")[1])/60)) 
            except:
                longitude.append(np.nan)            
```

```python
melbourne_stations = pd.DataFrame(columns=['Station',"Latitude",'Longitude'])
melbourne_stations['Station'] = stations_list
melbourne_stations['Latitude'] = latitude
melbourne_stations['Longitude'] = longitude
melbourne_stations
```

```python
melbourne_stations.to_csv(r'melbourne_stations.csv', index = False)
```

```python
melbourne_stations.dropna(inplace=True)
melbourne_stations.reset_index(inplace=True)
melbourne_stations.drop(['index'],inplace=True,axis=1)
melbourne_stations
```
